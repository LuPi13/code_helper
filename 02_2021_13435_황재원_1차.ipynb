#숫자 비교 게임

##숫자 비교 게임의 규칙은 다음과 같습니다.

### 1. 각 플레이어(사용자, 컴퓨터)는 0부터 10까지 11개의 숫자를 패에 가지고 시작합니다. 

### 2. 각 플레이어는 매 턴마다 서로 숫자를 하나씩 냅니다.
### 2-1. 숫자의 대소를 비교하여 더 큰 쪽은 1점을 얻습니다. 이후 사용한 패는 사라집니다.
### 2-2. 숫자가 서로 같을 경우 점수 계산 없이 서로 다시 패로 가져갑니다.
### 2-3. 단, 0은 10과 9를 이깁니다.

### 3. 모든 패를 사용한 후 점수가 더 높은 쪽이 승리합니다.
